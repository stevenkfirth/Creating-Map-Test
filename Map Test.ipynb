```python
df = pd.read_csv('GeoCdf2.csv')
df
```

```python
my_map = folium.Map(location=[51.753791, -0.330183], zoom_start=6)
my_map
```

```python
folium.Marker( 
        location=[51.753791, -0.330183],
        popup='Local Authority Name',
        tooltip='Local Authority Name'
    ).add_to(my_map)
```

```python
my_map = folium.Map(location=[51.753791, -0.330183], zoom_start=6)

for index, row in df.iterrows():
    folium.Marker( 
        location=[row['Latitude'], row['Longitude']],
        popup=row['Local Authority Name'],
        tooltip=row['Local Authority Name']
    ).add_to(my_map)
    
    if index==100: break  # this stops after the first 100 rows

my_map
```

```python
my_map = folium.Map(location=[51.753791, -0.330183], zoom_start=6)

for _, POSTCODE in df.iterrows():
    folium.Marker( 
        location=[df['Latitude'], df['Longitude']],
        popup=df['Local Authority Name'],
        tooltip=df['Local Authority Name']
    ).add_to(my_map)

my_map
```
